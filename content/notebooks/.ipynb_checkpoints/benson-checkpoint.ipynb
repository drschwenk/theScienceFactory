### Challenge 8

```python
def producePlot(timeSeries):
    station = '34 ST-HERALD SQ'
    dates = range(0,7)
    counts = timeSeries[station]
    pylab.plot(dates, [n/1000 for n in counts],linewidth=2)
    LABELS = ["Saturday","Sunday","Monday", "Tuesday", "Wednesday","Thursday","Friday"]
    plt.title('Weekly Subway Ridership for ' + station, fontsize=15)
    plt.xlabel("Day", fontsize=15, labelpad=10)
    plt.ylabel("Thousands of Riders", fontsize=15, labelpad=15)
    plt.xticks(dates,LABELS, rotation=60)
    return


fig, ax = plt.subplots()
for week in week_series:
	producePlot(week)

plt.show()
```

```python
historical_tot = defaultdict(int)
max_count = 1000000
for week in week_series:
    for station, counts in week.items():
        try:
            if sum(counts) > max_count or sum(counts) < 0:                  #Last chance sanity check on totals
                pass
            else:
                historical_tot[station] += sum(counts)
        except TypeError:    
            pass

```

```python
import operator
sorted_tots = sorted(historical_tot.items(), key=operator.itemgetter(1), reverse=True)
print(sorted_tots[:15])
```

```python
scaled_counts = [station[1]/float(10**6) for station in sorted_tots]

plt.hist(scaled_counts,bins=45)
plt.title('Frequency of High Traffic Stations', fontsize=17)
plt.xlabel("Millions of Riders", fontsize=15, labelpad=5)
plt.ylabel("Frequency", fontsize=15, labelpad=15)
plt.savefig('Total_freq.png')
```
